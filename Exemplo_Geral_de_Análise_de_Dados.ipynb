###2. Qual cidade com maior valor de venda de produtos na Categoria "Material  de Escritório"?


```python
dados2 = dados[dados['Categoria'] == 'Material de Escritório']
dados2 = dados2.groupby('Cidade').sum()
dados2  = dados2.sort_values('Valor_Venda', ascending=False)
dados2.head()
```
